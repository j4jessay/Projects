```python

import json
with open('C:\\Users\\Administrator\\ML\\data.txt') as f:
    lines = f.readlines()

data = []
for line in lines:
    data.append(json.loads(line))

df = pd.DataFrame(data)

df
```

```python
df1 = df.drop(['merchantName','acqCountry','merchantCountryCode','merchantCategoryCode','transactionDateTime','currentExpDate','accountOpenDate','dateOfLastAddressChange','merchantCity', 'merchantState', 'merchantZip','posOnPremises','recurringAuthInd','transactionType','echoBuffer'],axis = 'columns')
df1.head()
```

```python
df2 = df1.copy()
```

```python
df2["isFraud"] = df2["isFraud"].astype(int)
df2.head()
```

```python
df2 = df2.apply(pd.to_numeric,errors='coerce')
df2.info()
# errors{â€˜ignoreâ€™, â€˜raiseâ€™, â€˜coerceâ€™}, default â€˜raiseâ€™
# If â€˜raiseâ€™, then invalid parsing will raise an exception.

# If â€˜coerceâ€™, then invalid parsing will be set as NaN.

# If â€˜ignoreâ€™, then invalid parsing will return the input.
```

```python
df2.iloc[:] = df2.iloc[:].astype(float)
df2.info()
```

```python
df3 = df2.copy()
```

```python
df3 = pd.get_dummies(df2,
                     columns = ['cardPresent', 'expirationDateKeyInMatch'])
```

```python
df5 = df4.copy()
```

```python
df6 = df5.dropna()
```

```python
df6.isna().sum()
```

```python
y = df6['isFraud']
```

```python
x = df6.drop(['isFraud'],axis='columns')
```

```python
x.head()
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
clf = linear_model.LogisticRegression()
```

```python
clf.fit(x_train,y_train)
```

```python
import numpy as np
y_pred = np.array(clf.predict(x_test))
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
print(classification_report(y_test,y_pred))
```
